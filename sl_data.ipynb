```python
df = df.drop(columns =['GAME','PA','RBI','OBP','SLG'])
```

```python
df['LOCATION'] = df['LOCATION'].str.upper()
```

```python
df
```

```python
sns.barplot(data=df, x='YEAR', y='HOMERUN', hue='LOCATION')
sns.set_theme(rc={'figure.figsize':(12,6)}, style='white')
plt.title('SL HOMERUNS : HOME vs AWAY')
plt.xlabel('YEAR')
plt.ylabel('HOMERUN')
plt.show
```
